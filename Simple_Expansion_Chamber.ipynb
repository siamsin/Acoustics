# Basic Reactive Mufflers

## The Simple Expansion Chamber

The transmission loss of a simple expansion chamber is $$TL = 10log\{1+(\frac{m-1/m}{2}sinkl)^{2}\}$$ 
m is the area ratio given by $$m = {(\frac{D}{d})}^2$$
and l is the length of the muffler

```python
#Transmission Loss calculations
#all dimensions in m
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(5,2000,2000)
d=0.04                               #inlet and outlet pipe diameter
D=4*d                                #muffler diameter
m=(D/d)**2
L=0.4                                #length of muffler     
c=343                                #speed of sound
kn=2*np.pi*freq/c

TL=10*np.log10(1+(((m-1/m)/2)*np.sin(kn*L))**2);
#TL of Helmholtz resonator
plt.plot(freq,TL,'b')
plt.xlabel("frequency in Hz")
plt.ylabel("Transmission Loss in dB")
plt.show()
```
